```python
print("Кореляція")

dataDropna = data.drop(columns=['Year', 'Life expectancy ', 'Adult Mortality', 'Alcohol', 'under-five deaths '])
corelated = dataDropna.corr()
plot.figure(figsize=(12,18))

map = sb.diverging_palette(400, 150, s=80, l=50, n=10, center="light", as_cmap=True)
_ = sb.heatmap(corelated, center=0, annot=True, fmt='.3f', square=True, cmap=map)
```

```python
print('Дисперсія')
print('Life expectancy дисперсія: ', data["Life expectancy "].var())
print('Alcohol дисперсія: ', data["Alcohol"].var())
print('Hepatitis B дисперсія: ', data["Hepatitis B"].var())
print('Percentage expenditure дисперсія: ', data["percentage expenditure"].var())
```

```python
firstFactors = ['Adult Mortality','infant deaths','Alcohol',' BMI ','under-five deaths ','Population']
print("Дані, обрані спочатку")
for i in nmp.arange(0,6):
    corr = data[firstFactors[i]].corr(data['Life expectancy '])
    if(-0.09<=corr<=0 or 0<= corr<=0.09): print(firstFactors[i], " Кореляції немає ", corr)
    if(-0.3<corr<=-0.1or 0.1<=corr<0.3): print(firstFactors[i], " Низька кореляція", corr)
    if(-0.5<corr<=-0.3or 0.3<=corr<0.5): print(firstFactors[i], " Середня кореляція", corr)
    if(-1<=corr<=-0.5 or 0.5<=corr<=1): print(firstFactors[i], " Висока кореляція", corr)
allFactors = ['Adult Mortality', 'infant deaths','Alcohol','percentage expenditure','Hepatitis B','Measles ',' BMI ','under-five deaths ','Polio','Total expenditure','Diphtheria ',' HIV/AIDS','GDP','Population',' thinness  1-19 years',' thinness 5-9 years','Income composition of resources','Schooling']
print("Реальні тчинники що впливають")
for i in nmp.arange(0, 18):
    corr = data[allFactors[i]].corr(data['Life expectancy '])
    if(-0.5<corr<=-0.3or 0.3<=corr<0.5): print(allFactors[i], " Середня кореляція", corr)
    if(-1<=corr<=-0.5 or 0.5<=corr<=1): print(allFactors[i], " Висока кореляція", corr)
```

```python
print('Чи повинна країна з нижчим значенням очікуваної тривалості життя (<65) збільшити витрати на охорону здоров’я, щоб покращити середню тривалість життя?')
lessThan65 = data[data['Life expectancy ']<65]
lessThan65_example = lessThan65.sample(n = 20)
statistic = sb.scatterplot(lessThan65_example, x='Life expectancy ', y ='Total expenditure', hue = 'Country')
sb.move_legend(statistic, "upper left", bbox_to_anchor=(1, 1))
plot.show()

less65cor = lessThan65['Total expenditure'].corr(lessThan65['Life expectancy '])
print("Health care costs depend on the average life expectancy:")
if(-0.09<=less65cor<=0 or 0<= less65cor<=0.09): print(" Кореляція відсутня ", less65cor)

if(-0.3<less65cor<=-0.1or 0.1<=less65cor<0.3): print(" Низьку кореляцію.", less65cor)

if(-0.5<less65cor<=-0.3or 0.3<=less65cor<0.5): print( " Середню кореляцію", less65cor)

if(-1<=less65cor<=-0.5 or 0.5<=less65cor<=1): print( " Високу кореляцію", less65cor)
```

```python
print('Чи має очікувана тривалість життя позитивну чи негативну кореляцію зі звичками в їжі, способом життя, фізичними вправами, курінням, вживанням алкоголю тощо.')

alcoholCorr = data['Life expectancy '].corr(data['Alcohol'])
bmiCorr = data['Life expectancy '].corr(data[' BMI '])

if alcoholCorr >= 0 : print('Alcohol affects positive')
else: print('Alcohol affects negative')

if bmiCorr >= 0 : print('BMI affects positive')
else: print('BMI affects negative')
```

```python
print('Який вплив навчання в школі на тривалість життя людини?')

schoolingCorr = data['Life expectancy '].corr(data['Schooling'])
print('Affection schooling on life expectancy: ')
if(-0.9<=schoolingCorr<=0 or 0<= schoolingCorr<=0.09): print(" not affect", schoolingCorr)

if(-0.3<schoolingCorr<=-0.1or 0.1<=schoolingCorr<0.3): print("low affect", schoolingCorr)

if(-0.5<schoolingCorr<=-0.3or 0.3<=schoolingCorr<0.5): print("average affect", schoolingCorr)

if(-1<=schoolingCorr<=-0.5 or 0.5<=schoolingCorr<=1): print("high affect", schoolingCorr)
```

```python
print('Позитивний чи негативний зв’язок очікуваної тривалості життя з вживанням алкоголю?')

alcoholCorr = data['Life expectancy '].corr(data['Alcohol'])

if alcoholCorr >= 0 : print('Alcohol affects positive')
else: print('Alcohol affects negative')
```

```python
print('Чи зазвичай у густонаселених країнах тривалість життя нижча?')
sortedCountries = data.sort_values(by ='Population' ,ascending = False).head(50)
fig, x= plot.subplots()
x = sb.scatterplot(sortedCountries, x='Life expectancy ', y ='Population', hue = 'Country')
sb.move_legend(x, "upper left", bbox_to_anchor=(1, 1))
plot.show()
```

```python
print('Який вплив охоплення імунізацією на очікувану тривалість життя?')

schoolingCorr = data['Life expectancy '].corr(data['Schooling'])
print('Affection imunnization on life expectancy: ')
if(-0.9<=schoolingCorr<=0 or 0<= schoolingCorr<=0.09): print(" not affect", schoolingCorr)

if(-0.3<schoolingCorr<=-0.1or 0.1<=schoolingCorr<0.3): print("low affect", schoolingCorr)

if(-0.5<schoolingCorr<=-0.3or 0.3<=schoolingCorr<0.5): print("average affect", schoolingCorr)

if(-1<=schoolingCorr<=-0.5 or 0.5<=schoolingCorr<=1): print("high affect", schoolingCorr)
```
