```python
import pandas as pd
```

```python
user_col = ['UserID', 'Gender', 'Age', 'Occupation', 'Zip_code']
```

```python
users = pd.read_table("users.dat", sep = "::", engine ='python', names = user_col)
```

```python
users
```

```python
mov_col = ['MovieID', 'Title', 'Genres']
```

```python
movies = pd.read_table("movies.dat", sep = "::", engine = 'python', names = mov_col, encoding = "ISO-8859-1")
```

```python
movies
```

```python
rating_col = ['UserID','MovieID','Rating','Timestamp']
```

```python
ratings = pd.read_table("ratings.dat", sep = "::", engine = 'python', names = rating_col)
```

```python
ratings
```

```python
dataFrame = ratings.merge(users, how = 'inner', on = user_col[0]).merge(movies, how = 'inner', on = mov_col[0])
```

```python
frame_0_17 = dataFrame[dataFrame['Age'] == 1];
frame_18_24 = dataFrame[dataFrame['Age'] == 18];
frame_25_34 = dataFrame[dataFrame['Age'] == 25];
frame_35_44 = dataFrame[dataFrame['Age'] == 35];
frame_45_49 = dataFrame[dataFrame['Age'] == 45];
frame_50_55 = dataFrame[dataFrame['Age'] == 50];
frame_56_ = dataFrame[dataFrame['Age'] == 56];
```
