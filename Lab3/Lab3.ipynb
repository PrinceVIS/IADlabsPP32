```python

ld_columns=list(ld.columns)
covariance = np.empty(18, dtype=object)
idx = 0
print('COVARIANCE (life expectance):')
print('')
for i in np.arange(4, 22):
    f = ld_columns[i]
    covariance[idx] = ld['Life expectancy '].cov(ld[f])
    print(ld_columns[i],":",covariance[idx])
    idx=idx + 1

```

```python
correlation = np.empty(18, dtype=object)
idx = 0
print('CORRELATION:')
print('')
for i in np.arange(4, 22):
    f = ld_columns[i]
    correlation[idx] = ld['Life expectancy '].corr(ld[f])
    print(ld_columns[i],":",correlation[idx])
    idx=idx + 1
```

```python
dispersion = np.empty(18, dtype=object)
idx = 0
print('DISPERSION:')
print('')
for i in np.arange(4, 22):
    f = ld_columns[i]
    dispersion[idx] = ld[f].var()
    print(ld_columns[i],":",dispersion[idx])
    idx=idx + 1
```
