```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("resources/Life Expectancy Data.csv", header = 0)
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe(include='object')
```

```python
country_list = df['Country'].unique()
metrics_list = df.keys()
metrics_list
```

```python
for country in country_list:
    df.loc[df['Country'] == country, metrics_list] = df.loc[df['Country'] == country, metrics_list].interpolate()
df.dropna(inplace = True)
df.isnull().sum()
```

```python
df.info()
```

```python
pd.options.display.float_format = '{:.2f}'.format
df.describe().T
```

```python
sns.set(font_scale = 1.3)
data_numeric = df.select_dtypes(include = 'number')
column_numeric = data_numeric.columns.drop('Year')
```

```python
plt.figure(figsize = (24, 36))
plt.subplots_adjust(wspace = 0.3, hspace = None)

i = 1
for col_name in column_numeric:
    plt.subplot(7,3,i)
    sns.histplot(data = df[col_name],  bins = 25)
    sns.rugplot(data = df[col_name], height=-0.02, clip_on=False)
    i +=1
```

```python
plt.figure(figsize=(18,36))
plt.subplots_adjust(wspace=1, hspace=None)


i = 1
for col_name in column_numeric:
    plt.subplot(7,3,i)
    sns.boxplot(data=df, y=col_name)
    i +=1
```

```python
df.var()
```

```python
covariance = df.cov()
covariance
```

```python
plt.figure(figsize=(20,15))
sns.heatmap(df.corr(), annot=True, cmap='bwr')
plt.show()
```

```python
plt.figure(figsize=(20,15))
sns.heatmap(df.corr(method="spearman"), annot=True, cmap='bwr')
plt.show()
```
