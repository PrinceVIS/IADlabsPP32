```python
column_names = list(dataset.columns)
offset = 4

j=0
cov = np.zeros(len(column_names) - 4)
for i in np.arange(offset, len(column_names)):
    factor = column_names[i]
    cov[j] = dataset['Life expectancy '].cov(dataset[factor])
    print("Covariation of Life expectancy ", factor, ":", cov[j])
    j+=1
```

```python
j = 0
cor_coef = np.zeros(len(column_names) - 4)
for i in np.arange(offset, len(column_names)):
    factor = column_names[i]
    cor_coef[j] = dataset['Life expectancy '].corr(dataset[factor])
    print("Correlation ", factor, cor_coef[j])
    j += 1
```

```python
j=0
var = np.zeros(len(column_names) - 4)
for i in np.arange(offset, len(column_names)):
    factor = column_names[i]
    var[j] = dataset[factor].var()
    print("Variance ", factor, var[j])
    j+=1
```

```python
factors = column_names[4 : len(column_names) : 1]

data = zip(factors, cov, cor_coef, var)
complite_results = ['Factor','Covariation','Correlation','Variance']
results = pd.DataFrame(data=data, columns = complite_results)
print(results)
```

```python
sns.set_theme(style='whitegrid', rc={'axes.facecolor': sns.xkcd_rgb['white'], 'figure.facecolor': sns.xkcd_rgb['steel']})
sns.catplot(results, y='Correlation', hue='Factor')
```
